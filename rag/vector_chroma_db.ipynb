```python
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
import os

print("üóÑÔ∏è Initializing ChromaDB Vector Database")
print("=" * 50)

# ==============================
# Create Persistent Chroma Client
# ==============================

persist_directory = "/content/drive/MyDrive/mcp and rag/chroma_DB"

print("1Ô∏è‚É£ Creating persistent ChromaDB client...")

client = chromadb.Client(
    Settings(
        persist_directory=persist_directory,
        is_persistent=True
    )
)

print("   ‚úÖ ChromaDB client created")

# ==============================
# Create Collection
# ==============================

print("2Ô∏è‚É£ Creating collection for TechCorp documents...")

collection = client.get_or_create_collection(name="techcorp_docs")

print("   ‚úÖ Collection 'techcorp_docs' ready")

# ==============================
# Load Embedding Model
# ==============================

print("3Ô∏è‚É£ Loading embedding model...")

model = SentenceTransformer('all-MiniLM-L6-v2')

dimension = model.get_sentence_embedding_dimension()

print(f"   ‚úÖ Model loaded: {dimension} dimensions")

# ==============================
# Test Document
# ==============================

print("4Ô∏è‚É£ Testing with sample document...")

test_doc = "TechCorp allows remote work up to 3 days per week"

test_embedding = model.encode([test_doc])

print(f"   ‚úÖ Sample embedding created: {len(test_embedding[0])} dimensions")

# ==============================
# Add Document to Collection
# ==============================

collection.add(
    documents=[test_doc],
    embeddings=test_embedding.tolist(),
    ids=["test_doc_1"]
)

print("   ‚úÖ Test document added to collection")

# ==============================
# Verify Collection
# ==============================

print("5Ô∏è‚É£ Verifying collection...")

count = collection.count()

print(f"   ‚úÖ Collection contains {count} documents")

print()
print("üéâ ChromaDB Vector Database Initialized Successfully!")
print(f"üìä Collection: techcorp_docs")
print(f"üìä Embedding dimensions: {dimension}")
print(f"üìä Documents stored: {count}")

# ==============================
# Create Completion Marker in Drive
# ==============================

marker_path = "/content/drive/MyDrive/mcp and rag/chroma_DB/chroma_db_initialized.txt"

with open(marker_path, "w") as f:
    f.write("ChromaDB vector database initialized successfully")

print("‚úÖ Initialization complete!")
```

```python
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
import os

print("üìö Storing TechCorp Documents in Vector Database")
print("=" * 50)

# ==============================
# Connect to Existing Persistent DB
# ==============================

persist_directory = "/content/drive/MyDrive/mcp and rag/chroma_DB"

client = chromadb.Client(
    Settings(
        persist_directory=persist_directory,
        is_persistent=True
    )
)

collection = client.get_or_create_collection("techcorp_docs")

model = SentenceTransformer('all-MiniLM-L6-v2')

print("   ‚úÖ Connected to persistent DB")

# ==============================
# Load Documents From Your Folder
# ==============================

folder_path = "/content/drive/MyDrive/mcp and rag/database"

docs = []
doc_paths = []

for file in os.listdir(folder_path):
    if file.endswith(".md"):
        full_path = os.path.join(folder_path, file)
        with open(full_path, "r", encoding="utf-8") as f:
            docs.append(f.read())
            doc_paths.append(full_path)

print(f"   ‚úÖ Loaded {len(docs)} documents")

# ==============================
# Create Embeddings
# ==============================

print("3Ô∏è‚É£ Creating embeddings...")
embeddings = model.encode(docs)
print(f"   ‚úÖ Created {len(embeddings)} embeddings")

# ==============================
# Generate Unique IDs
# ==============================

doc_ids = [f"doc_{i+1}" for i in range(len(docs))]

# ==============================
# Store in ChromaDB
# ==============================

print("4Ô∏è‚É£ Storing documents...")

collection.add(
    documents=docs,
    embeddings=embeddings.tolist(),
    ids=doc_ids
)

print(f"   ‚úÖ Stored {len(docs)} documents")

# ==============================
# Verify Storage
# ==============================

count = collection.count()
print(f"   ‚úÖ Vector database now contains {count} documents")

print("\nüéâ Documents Successfully Stored!")
```

```python
query = "remote work policy"

query_embedding = model.encode([query])

results = collection.query(
    query_embeddings=query_embedding.tolist(),
    n_results=3
)

print("üîé Query:", query)
print("üìÑ Retrieved Documents:")
for doc in results["documents"][0]:
    print("-", doc[:150], "...\n")
```

```python
def semantic_search(query, top_k=3):
    query_embedding = model.encode([query])
    results = collection.query(
        query_embeddings=query_embedding.tolist(),
        n_results=top_k
    )
    return results["documents"][0]
```
